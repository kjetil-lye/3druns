```python
import sys
sys.path.append('../python')

import plot_info
import convergence
import os
import os.path
```

```python
KH_DATA_PATH = plot_info.get_environment("THREEDIMS_PATH", [])
perturbations = [0.1, 0.01]
resolutions = [64, 128, 256, 512]

statistics = ["mean", "variance"]
variable = "rho"

for perturbation in perturbations:
    basename = os.path.join(KH_DATA_PATH, 'kelvinhelmholtz_3d_tube/stats/kelvinhelmholtz_{resolution}_' + str(perturbation) + "/kh_{statistic}_1.nc")
    setup = "Kelvin-Helmholtz $\\epsilon={}$".format(perturbation)
    convergence.plot_statistics_convergence(resolutions, basename, statistics, variable, setup)
    basename_single = os.path.join(KH_DATA_PATH, 'kelvinhelmholtz_3d_tube/stats/kelvinhelmholtz_{resolution}_' + str(perturbation) + "/kh_1.nc")
    try:
        convergence.plot_single_sample_convergence(resolutions, basename_single, 0, variable, setup)
    except:
        print("Failed plotting single sample convergence for perturabtion = {}".format(perturbation))
```
